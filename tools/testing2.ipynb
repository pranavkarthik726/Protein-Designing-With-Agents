```python
import os

import Bio.SeqIO
import pymuscle5

path = r"input.fasta"
records = list(Bio.SeqIO.parse(path, "fasta"))

sequences = [
    pymuscle5.Sequence(record.id.encode(), bytes(record.seq))
    for record in records
]

aligner = pymuscle5.Aligner()
msa = aligner.align(sequences)

for seq in msa.sequences:
    print(f">{seq.name.decode()}")
    print(seq.sequence.decode())
```

```python
from Bio import AlignIO

# Load your MUSCLE alignment
alignment = AlignIO.read("input.aln", "fasta")

# Scan for conserved columns
motif = ""
threshold = 3  # Total number of sequences

for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    residues = [res for res in column if res != '-']
    if len(residues) >= threshold and len(set(residues)) == 1:
        motif += residues[0]
    else:
        motif += "."

# Print conserved motif pattern
print(motif)
```

```python
import re

# Using the string you got, e.g.:
conserved_line = ".....................................................................................................................................................................................................................................................................................................................................................................................................................P............................................................................................................................................................................................................................................................................................................................................................................................................P.........................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................................."
# Define a pattern for stretches of conserved letters (3 or more in a row)
conserved_blocks = re.finditer(r"[A-Z]{3,}", conserved_line)

# Print positions and motifs
for match in conserved_blocks:
    start = match.start()
    end = match.end()
    motif = match.group()
    print(f"Conserved region from {start} to {end}: {motif}")
```
